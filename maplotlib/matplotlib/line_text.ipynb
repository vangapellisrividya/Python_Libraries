```python
'''
    Created on Dec 5, 2021

    @author:srividya
    @Date:  5/12/2021
    @Title: Draw line by given textfile

    '''

import matplotlib.pyplot as plt
with open("test.txt") as f:
    data = f.read()
data = data.split('\n')
x = [row.split(' ')[0] for row in data]
y = [row.split(' ')[1] for row in data]
plt.plot(x, y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Sample graph!')
plt.show()
```
